```python
import matplotlib.pyplot as Collo
import numpy as kolosh

# Data from the image
months = ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019']
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social_media = [70, 80, 90, 87, 92]

x = kolosh.arange(len(months))  # Locate months in the X-axis
width = 0.26  # Bar width for spacing purposes btwn each group bar

fig, ax = Collo.subplots(figsize=(8, 4))# Graph customaization (8 = length and 4 = height)

# Plotting bars
ax.bar(x - width, searches, width, label='Searches', color='blue')
ax.bar(x, direct, width, label='Direct', color='pink')
ax.bar(x + width, social_media, width, label='Social Media', color='orange')

# Labels and formatting
ax.set_xlabel('months', loc='right')  # Moves the label to the right
ax.set_ylabel('visitors\nin thousands')
ax.set_title('Visitors by web traffic sources')
ax.set_xticks(x)
ax.set_xticklabels(months)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, frameon=False)# Remove legend border
ax.set_ylim(0, 100)  # Force y-axis to go from 0 to 100

Collo.show()
```
